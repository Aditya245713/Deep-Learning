# Question 01: [ Marks 10 ]

What is a perceptron?
Explain its three main components and their roles.

**Write** Answer 01:
Perceptron is a specialized learning algorithm that is used for binary classification. It classifies data into one of two groups based on the threshold that has been set.
Three main components of perceptron are:

i. Weights(Determines the impact of input)

ii. Bias(Shifts decision boundary)

iii. Activation function(groups data based on threshold)

# Question 02: [ Marks 10 ]

What is a decision boundary in a perceptron?
Explain how the equation
w₁x₁ + w₂x₂ + b = 0
represents a decision boundary geometrically.

## Write Answer 02:
The decision boundary is the line that separates the data of two different groups.

The equation w₁x₁ + w₂x₂ + b = 0 is the equation of a straight line that separates the two classes of the binary classification. Any value greater than this falls in the positive class, and the values less than this falls in the negative class. w₁ and w₂ determines the slope of separation, and bias b moves the line up or down, left or right.

# Question 03: [ Marks 15 ]

Why is the perceptron called a linear classifier?
Explain your answer using the concept of linear separability and logical gate examples such as AND, OR, and XOR.

## Write Answer 03:
Perceptron is a linear classifier as it linearly classifies the data into two classes.

Linear separability means the data of two groups are separable by a straight line. In case of logic gates:

AND Gate: Outputs are linearly separable and Perceptron works on it properly.

OR Gate: Outputs are linearly separable and Perceptron works on it properly.

XOR Gate: Outputs are not separable by a straight line and perceptron fails on it too.

Therefore, perceptron is a linear classifier.

# Question 05: [ Marks 10 ]

## Write Answer 05:
XOR problem outputs are not separable linearly, hence a single-layer perceptron fails to solve it.

This limitation states that not every problem is solvable through simple linearly separable neural networks, which is why we solemnly need multilayer neural networks, so that we can solve complex problems.
